```python
import random
from collections import deque
import copy

'''
Funcion que inicializa la poblacion inicial en base al alfabeto
'''
def create (alphabet, N=100): # crea y devuelve población
    pop = []
    for _ in range(N):
        individual = random.sample(alphabet, len(alphabet))  # Permutación aleatoria
        pop.append(individual)
    return pop

'''
Funcion que, dado una poblacion y la funcion fitness, retorna una tupla 
(poblacion de individuos ordenada, fitness de los individuos)
'''
def sort_pop (pop, fitness): # devuelve una tupla: la población ordenada por fitness, y la lista de fitness

    pop_with_fitness = [(indiv, fitness(indiv)) for indiv in pop]
    sorted_pop = sorted(pop_with_fitness, key=lambda x: x[1]) # x[1] es el fitness para cada individuo
    #print(f"ORDEN {[fit for _, fit in sorted_pop]}")
    return [indiv for indiv, _ in sorted_pop], [fit for _, fit in sorted_pop]


''' 
Seleccion por torneo: Se eligen T individuos de la población (por ejemplo T=2) de
modo equiprobable (todos los individuos tienen la misma probabilidad de ser escogidos), 
y se añade a la nueva población el de mayor calidad.
'''
def select (pop, T): # devuelve un individuo seleccionado por torneo, devuelve una copia para evitar efectos laterales. pop se supone ya ordenada por fitness  
    tournament = random.sample(pop, T)  
    return max(tournament, key=lambda ind: fit_cities(ind)).copy()


'''
En los cromosomas en los que no sepuede repetir un valor, como en el TSP 
(donde una ciudad no debe aparecer dos veces en un cromosoma), es necesario aplicar operadores para 
permutacion: se basa en elegir N posiciones fijas el padre _a_ y copiarlas al hijo _a'_, luego se rellenan
los huecos tomando del padre _b_ los elementos que no estén en _a'_ y se añaden de modo ordenado a los huecos.
Igual para _b'_

Se llama Crossover con máscara de orden.

    - Param pcross: determina si se cruzan los padres para generar un hijo, o por el contrario se devuelven 
                    los padres originales
'''
def crossover (ind1, ind2, pcross): # devuelve el cruce (emparejamiento) de dos individuos
    if random.random() > pcross:    # Si pcross = 0.70, la probabilidad de que se genere un numero entre 0.7 y 1 es de 0.3.
        return ind1.copy(), ind2.copy()

    size = len(ind1)
    start, end = sorted(random.sample(range(size), 2))  # Extremos de la mascara

   
    child1 = [-1] * size
    child1[start:end] = ind1[start:end]

    child2 = [-1] * size
    child2[start:end] = ind2[start:end]

    
    def fill_child(child, parent):
        gene_nonRepeated = deque()  # Almacena los genes del padre que no existen en el hijo

        for gene in parent: # gene_nonRepeated contiene los genes de parent que no estan en child
            if gene not in child:
                gene_nonRepeated.append(gene)  # se añade al final de la deque

        # se asignan los genes no repetidos a las posiciones vacías en child
        for idx in range(len(child)):  # para cada índice en el hijo
            if child[idx] == -1:  # si el gen es -1, necesitamos llenarlo
                child[idx] = gene_nonRepeated.popleft()  # desencola primer gen de gene_nonRepeated 

        return child

    child1 = fill_child(child1, ind2)
    child2 = fill_child(child2, ind1)

    return child1, child2

'''
Mutacion equiprobable 1/longitud_cromosoma para cada uno de los genes de un individuo
'''
def mutate (ind, pmut): # devuelve individuo mutado; la mutación consistirá en intercambiar elementos

    for i in range(len(ind)):
        if np.random.rand() < pmut: # se escoge un indice aleatorio con el que se va a reemplazar el gen "i"
            i_2 = np.random.randint(0, len(ind))
            ind[i], ind[i_2] = ind[i_2], ind[i]
    return ind


'''
Algoritmo de evolución generacional
'''
def evolve (pop, fit, pmut, pcross=0.7, ngen=100, T=2, trace=0, elitism=False):

    pop, _ = sort_pop(pop, fit)

    for i in range(ngen):
        new_poblacion = []

        if(elitism):
            best_indv = pop[0]
            new_poblacion.append(best_indv.copy())

        while len(new_poblacion) < len(pop):
            parent_1 = select(pop, T)
            parent_2 = select(pop, T)

            child_1, child_2 = crossover(parent_1, parent_2, pcross)

            child_1 = mutate(child_1, pmut)
            child_2 = mutate(child_2, pmut)

            new_poblacion.extend([child_1, child_2])
        
        pop, fitness = sort_pop(new_poblacion[:len(pop)], fit)

        if i % trace == 0 and trace > 0:
            print(f"Generacion {i}: mejor fitness [{fitness[0]}]")
    
    print(f"Generacion {ngen}: mejor fitness [{fitness[0]}]")

    return pop
```

```python
# Función himmelblau
# si se pone onlyone a True 
import random
# Esta funcion recibe un solo cromosoma
onlyone = False
def himmelblau (ch):
	x = ch[0]
	y = ch[1]
	fxy = (x**2 + y - 11)**2 + (x + y**2 -7)**2
	if onlyone and (x>0 or y>0): # himmelblau modified to have only one global minimum (-3.77, -3.28)
		fxy += 0.5
	return fxy
```

```python
# fitness para himmelblau: valor mínimo de la función, a mayor valor de fitness, menor valor de la función
#maximizamos función

def fitness_himmel (ch):
	return 1 / (1 + himmelblau(ch))
```

```python
# Define operadores de números reales

rang = (-5, 5) # al no hacerlo con clases, debemos definir el rango como variable global

def select (pop, T): # devuelve un individuo seleccionado por torneo, devuelve una copia para evitar efectos laterales. pop se supone ya ordenada por fitness  
    tournament = random.sample(pop, T)
    # Selección por torneo: selecciona el individuo con el mayor fitness (es decir, el más bajo en valor de Himmelblau)  
    return max(tournament, key=lambda ind: fitness_himmel(ind)).copy()

def create (alphabet=None, N=100): # crea y devuelve población; alphabet aquí no se usa
    pop = []

    for _ in range(N):
        values = [random.uniform(-5, 5) for _ in range(2)]  
        pop.append(values)
    return pop

def sort_pop (pop, fitness): # devuelve una tupla: la población ordenada por fitness, y la lista de fitness
    pop_with_fitness = [(indiv, fitness(indiv)) for indiv in pop]
    sorted_pop = sorted(pop_with_fitness, key=lambda x: x[1], reverse=True)  # Mayor fitness primero
    # print(f"ORDEN {[fit for _, fit in sorted_pop]}")
    return [indiv for indiv, _ in sorted_pop], [fit for _, fit in sorted_pop]

def crossover (ind1, ind2, pcross): # devuelve el cruce (emparejamiento) de dos individuos
    
    if (random.random() > pcross):
        return ind1.copy(), ind2.copy()

    beta = random.uniform(0,1)
    
    child1_x = beta*ind1[0] + (1 - beta)*ind2[0]
    child1_y = beta*ind1[1] + (1 - beta)*ind2[1]

    child2_x = (1 - beta)*ind1[0] + beta*ind2[0]
    child2_y = (1 - beta)*ind1[1] + beta*ind2[1]


    return [child1_x, child1_y], [child2_x, child2_y]

def mutate(ind, pmut):
    #individual = random.sample(rang, 2) 
    if random.random() < pmut:
        ind[random.randint(0, 1)] = random.uniform(rang[0], rang[1]) # otro tipo de mutacion
        #ind[0] = individual[0]
        #ind[1] = individual[1]
    return ind.copy()

'''
Algoritmo de evolución generacional
'''
def evolve (pop, fit, pmut, ngen=100, T=2, trace=0, pcross=0.7, elitism=False):

    pop, _ = sort_pop(pop, fit)

    for i in range(ngen):
        new_poblacion = []

        if(elitism):
            best_indv = pop[0]
            new_poblacion.append(best_indv.copy())
            
        while len(new_poblacion) < len(pop):
            parent_1 = select(pop, T)
            parent_2 = select(pop, T)

            child_1, child_2 = crossover(parent_1, parent_2, pcross)

            child_1 = mutate(child_1, pmut)
            child_2 = mutate(child_2, pmut)

            new_poblacion.extend([child_1, child_2])
        
        pop, fitness = sort_pop(new_poblacion[:len(pop)], fit)

        if i % trace == 0 and trace > 0:
            print(f"Generacion {i}: mejor fitness [{fitness[0]}]")
    
    print(f"Generacion {ngen}: mejor fitness [{fitness[0]}]")

    return pop
```

```python
'''PRUEBA EN NUEVA GENERACIÓN'''
#Algoritmos meméticos (diap 78), mezclar exploración de genéticos con técnicas de optimización de mates
''' Búsqueda local (descenso por el gradiente) de hijos antes de definir la nueva generació'''
def local_search(individual, step_size=0.1, max_iterations=1):
    x, y = individual
    for _ in range(max_iterations):
        # Derivadas parciales de la función Himmelblau
        grad_x = 4*x*(x**2 + y - 11) + 2*(x + y**2 - 7)
        grad_y = 2*(x**2 + y - 11) + 4*y*(x + y**2 - 7)
        
        # Nuevos valores dell individuo
        x -= step_size * grad_x
        y -= step_size * grad_y

        x = max(rang[0], min(rang[1], x))
        y = max(rang[0], min(rang[1], y))
    
    return [x, y] 


# Aplicar a evolve (el resto del código se mantiene pero se aplica la nueva forma de seleccionar la siguiente generación)
def evolve(pop, fit, pmut, ngen=100, T=2, trace=0, pcross=0.7, elitism=False, 
           step_size=0.1, local_search_prob=0.1, ls_function=None):
    """
        step_size: Tamaño del paso para búsqueda local.
        local_search_prob: Probabilidad de aplicar búsqueda local.
        ls_function: Función de búsqueda local, por si se quiere cambiar.
    """
    pop, _ = sort_pop(pop, fit)

    for i in range(ngen):
        new_poblacion = []

        if elitism:
            best_indv = pop[0]
            new_poblacion.append(best_indv.copy())
            
        while len(new_poblacion) < len(pop):
            parent_1 = select(pop, T)
            parent_2 = select(pop, T)

            child_1, child_2 = crossover(parent_1, parent_2, pcross)

            child_1 = mutate(child_1, pmut)
            child_2 = mutate(child_2, pmut)

            # Nuevo: Aplicar búsqueda local si se selecciona
            if ls_function and random.random() < local_search_prob:
                child_1 = ls_function(child_1, step_size)
            if ls_function and random.random() < local_search_prob:
                child_2 = ls_function(child_2, step_size)

            new_poblacion.extend([child_1, child_2])
        
        pop, fitness = sort_pop(new_poblacion[:len(pop)], fit)

        if i % trace == 0 and trace > 0:
            print(f"Generacion {i}: mejor fitness [{fitness[0]}]")
    
    print(f"Generacion {ngen}: mejor fitness [{fitness[0]}]")

    return pop




#Visualizamos el resultado
population = create()
population = evolve(population, fit=fitness_himmel, pmut=0.1, ngen=1000, T=3, trace=100,elitism=True, step_size=0.1, local_search_prob=0.1, ls_function=local_search)
best_individual=population[0]
himmel_value = himmelblau(best_individual)

print(f"El mejor individuo es {best_individual}")
print(f"Valor de la función de Himmelblau en ese individuo: {himmel_value:.10f}")
print(f"Valor de la función de Himmelblau: {himmelblau(best_individual)}")

```

```python
'''PRUEBA EN MUTACIÓN HIJOS'''

#Búsqueda más elaborada para mutar los hijos, antes era aleatoria la mutación

def gradient(ch, epsilon=1e-5):
    x, y = ch[0], ch[1]
    # Derivadas parciales de los genes x e y 
    grad_x = (himmelblau([x + epsilon, y]) - himmelblau([x, y])) / epsilon
    grad_y = (himmelblau([x, y + epsilon]) - himmelblau([x, y])) / epsilon
    
    return grad_x, grad_y

# Mutación basada en el gradiente
def mutate_with_gradient(individual, pmut, learning_rate=0.1, epsilon=1e-5):
    if random.random() < pmut:
        # Gradientes
        grad_x, grad_y = gradient(individual, epsilon)
        
        # Actualizar en la dirección opuesta al gradiente x_t+1= x_t - alpha*gradiente
        new_x = individual[0] - learning_rate * grad_x
        new_y = individual[1] - learning_rate * grad_y

        new_x = max(rang[0], min(rang[1], new_x))
        new_y = max(rang[0], min(rang[1], new_y))
        
        return [new_x, new_y]

    # Si no se muta, se mantiene igual
    return individual

# Función evolve probando la mutación
def evolve(pop, fit, pmut, ngen=100, T=2, trace=0, pcross=0.7, elitism=False):

    pop, _ = sort_pop(pop, fit)

    for i in range(ngen):
        new_poblacion = []

        if(elitism):
            best_indv = pop[0]
            new_poblacion.append(best_indv.copy())
            
        while len(new_poblacion) < len(pop):
            parent_1 = select(pop, T)
            parent_2 = select(pop, T)

            child_1, child_2 = crossover(parent_1, parent_2, pcross)

            # Usar mutación basada en el gradiente
            #print('mutación')
            child_1 = mutate_with_gradient(child_1, pmut)
            child_2 = mutate_with_gradient(child_2, pmut)

            new_poblacion.extend([child_1, child_2])
        
        pop, fitness = sort_pop(new_poblacion[:len(pop)], fit)

        if i % trace == 0 and trace > 0:
            print(f"Generacion {i}: mejor fitness [{fitness[0]}]")
    
    print(f"Generacion {ngen}: mejor fitness [{fitness[0]}]")

    return pop

#Visualizamos el resultado
population = create()
population = evolve(population, fit=fitness_himmel, pmut=0.1, ngen=1000, T=3, trace=100, pcross=0.7, elitism=True)
best_individual=population[0]
himmel_value = himmelblau(best_individual)

print(f"El mejor individuo es {best_individual}")
print(f"Valor de la función de Himmelblau en ese individuo: {himmel_value:.10f}")
print(f"Valor de la función de fitness en ese individuo: {fitness_himmel(best_individual)}")
```
